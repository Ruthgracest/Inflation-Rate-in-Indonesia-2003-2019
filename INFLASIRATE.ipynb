```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```
df=pd.read_csv("/content/drive/My Drive/App/Inflasi - Inflasi (2).csv")
```

##**Data Preparation**##

```
df['Month'] = pd.to_datetime(df['Month']).dt.strftime('%b %Y')
```

```
df['Year'] = pd.DatetimeIndex(df['Month']).year
```

```
df['Month']=df.Month.apply(lambda x:x.split(' ')[0])
```

```
import re
df2 = df.drop(np.flatnonzero(df['Inflasi'].str.contains('--', na=False)))
number_re = re.compile(r'\s|<|%|>')
str2float = lambda x: x if np.isreal(x) else number_re.sub('', x, re.U)
df['Inflasi'] = df['Inflasi'].map(str2float).astype(float)
```

```
df.rename(columns={'Inflasi': 'Inflasi (%)'}, inplace=True)
```

##**Inflation Rate in Indonesia**

#2003-2006

```
cats = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df['Month'] = pd.Categorical((df['Month']), ordered=True, categories=cats)
plt.figure(figsize=(30,10))
ax=sns.lineplot(data=df[(df.Year.isin([2003,2004,2005,2006]))],x='Month',y='Inflasi (%)', hue='Year').set(title='Nilai Inflasi tahun 2003-2006')
```

Tahun **2005** inflasi di Indonesia meningkat drastis menyentuh angka **18%**, hal tersebut terjadi karena saat itu SBY memutuskan memangkas subsidi BBM lewat peningkatan harga BBM  lebih dari dua kali lipat. Angka inflasi di Indonesia berada di antara **14%-19%** hingga **Oktober 2006**. Hal itu dikarenakan hanya minyak internasional sedang meningkat pesat.

#2007-2010

```
cats = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df['Month'] = pd.Categorical((df['Month']), ordered=True, categories=cats)
plt.figure(figsize=(30,10))
ax = sns.lineplot(data=df[(df.Year.isin([2007,2008,2009,2010]))],x='Month',y='Inflasi (%)', hue='Year', palette='viridis').set(title='Nilai Inflasi tahun 2007-2010')
```

Tidak banyak terjadi perubahan besar inflasi dari akhir tahun 2006-2007. Namun pada tahun 2008 harga minyak dunia berada dipuncaknya yang menyebabkan harga BBM di Indonesia kembali naik dan besar inflasi Indonesia kembali melonjak ditambah lagi terjadi demonstrasi-demonstrasi di Indonesia saat itu. **Tahun 2009** nilai inflasi di Indonesia sangat turun rata-rata **3%-5%** karena harga BBM juga sudah diturunkan SBY mengingat hal itu merupakan salah satu strategi karena akan maju untuk Pemilu kedua.
